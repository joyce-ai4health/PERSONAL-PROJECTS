```python
# ðŸ§® Project 2: The Discount Calculator
# ðŸŽ¯ Goal:Create a program that:

# 1. Asks the user for the original price of an item

# 2. Asks the user for the discount percentage

# 3. Calculates the discount amount

# Displays: The discount amount and The final price after discount
```

```python
original_price = float(input("How much is the Item(s) ? $" ))
discount_percentage = float(input("what is the dicount percentage on the item(s) ? %  "))
discount_amount = original_price * (discount_percentage / 100)
final_price = original_price - discount_amount
print("The discount amount on the item(s): $ " + str(discount_amount))
print("The final price to pay:  $ " + str(final_price))
```

```python
# ðŸŽ“ Project 6: Student Grade Calculator
# Instructions:Ask the user for scores in 5 subjects

# Calculate the average score

# Display:

# The average score

# The grade based on the average:

# 70 â€“ 100 â†’ A

# 60 â€“ 69 â†’ B

# 50 â€“ 59 â†’ C

# 45 â€“ 49 â†’ D

# Below 45 â†’ F
```

```python
scores = []

for i in range(1, 6):
    score = int(input(f"Enter your score for subject {i}: "))
    scores.append(score)

average_score = sum(scores) / len(scores)
print("Your scores:", scores)
print(f"Average score: {average_score:.2f}")
# f â†’ â€œfloating-point numberâ€ (decimal number),.2 â†’ â€œshow 2 digits after the decimal pointâ€
if average_score >= 70:
    print("Grade = A (Excellent performance)")
elif average_score >= 69:
    print("Grade = B (Good performance)")
elif average_score >= 59:
    print("Grade = C (Average performance)")
elif average_score >= 45:
    print("Grade = D (Below average)")
else:
    print("Grade = F (Fail)")


```

```python
# ðŸ§® Project 7: Exam Result Checker
# ðŸŽ¯ Goal:Create a program that:

# Asks the user for marks in 6 subjects

# 1. Calculates the total marks and percentage

# 2.Determines pass/fail for each subject (pass â‰¥ 50)

# 3.Prints a final remark based on the overall percentage: â‰¥ 70% â†’ Distinction, 60â€“69% â†’ Merit, 50â€“59% â†’ Pass, < 50% â†’ Fail
```

```python
marks = []
for i in range(1,7):
    mark = int(input(f"Enter your mark  {i}: "))
    marks.append(mark)
    total_marks = sum(marks)
    percentage = (total_marks / 600) * 100

print("Your marks:", marks)
print(f"Total marks: {total_marks:.2f}")
print(f"Percentage: {percentage:.2f} % ")
# f â†’ â€œfloating-point numberâ€ (decimal number),.2 â†’ â€œshow 2 digits after the decimal pointâ€
if percentage >= 70:
    print(" Distinction")

elif percentage >= 69:
    print("Merit")
elif percentage >= 59:
    print("Pass")

else:
    print("Grade = F (Fail)")

```

```python
# Project 7: The Odd/Even Checker
#  The goal: is to create a programme that asks users for any number and intelligently determines whether it's odd or even.
# he Mathematical Secret: We'll use the modulus operator % that we learned earlier. Here's the key insight:
# when you divide an even number by 2, there's no remainder (it divides perfectly).
# When you divide an odd number by 2, there's always 1 left over.

```

```python
number = int(input("pick a number:"))
if number % 2 == 0:
   print("It is an even number")
else:
   print("it is an odd number")
```
